```python
import numpy as np
arr1 = np.random.randint(11, size = (1,10))
print(f'Матрица 1: {arr1}')
arr2 = np.random.randint(11, size = (1,10))
print(f'Матрица 2: {arr2}')
arr3 = arr1+arr2
print(f'Сумма матриц: {arr3}')
arr3 = arr3.reshape((2,5))
print(arr3)
print(arr3.mean())
```

```python
arr1 = np.random.randint(15, size = (3,3))
print(arr1[0])
print(arr1[:,0:1])
arr1 = arr1.reshape(9,1)
print(arr1)
arr1 = arr1.flatten()
print(arr1)
```

```python
arr1 = np.random.randint(1, 11, size=(4, 3))
print(arr1)
result = np.vstack((arr1[:, [0]], arr1[:, [1]], arr1[:, [2]]))
print(result)
```

```python
arr1 = np.random.randint(1,31, size = 9)
print(arr1)
max_index = np.argmax(arr1)
print("Индекс максимального элемента:", max_index)
print("Максимальное значение:", arr1[max_index])
```

```python
arr1 = np.random.randint(1,6, size = (2,2))
arr2 = np.random.randint(1,6, size = (2,2))
print(f'Матрица 1: {arr1}')
print(f'Матрица 2: {arr2}')
tr = np.transpose(arr1)
print(f'Транспонированная матрица: {tr}')
print(f'Сложение матриц: {arr1 + arr2}')
```
