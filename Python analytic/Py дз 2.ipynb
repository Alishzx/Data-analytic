```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

image = Image.open("imagea.jpg")
image = np.array(image)
blue_white_image = np.copy(image)

blue_white_image[:, :, 0] = 0
blue_white_image[:, :, 1] = 0
bright_areas = blue_white_image[:, :, 2] > 120
blue_white_image[bright_areas, 2] = 255

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")
plt.subplot(1, 2, 2)
plt.imshow(blue_white_image)
plt.title("Blue and White Image")
plt.axis("off")

plt.show()
```
