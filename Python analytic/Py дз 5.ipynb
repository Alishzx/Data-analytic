```python
import pandas as pd
df = pd.read_csv('ДЗ_Pandas.csv')
df.head()
df1 = df.dropna(subset = ['Бюджет (млн $)'] , inplace = True)
print(df1)
df1 = df.fillna({'Режиссер': 'Неизвестен'}, inplace = True)
print(df1)
df1 = df.fillna({'Рейтинг': df['Рейтинг'].mean()}, inplace=True)
print(df1)
```

```python
df1 = df.groupby('Жанр')[['Рейтинг', 'Бюджет (млн $)']].mean()
print(df1)
df1 = df.groupby('Год выпуска').agg({'Название фильма': 'count', 'Сборы (млн $)' : 'sum'})
print(df1)
df1 = df.groupby(['Жанр', 'Год выпуска']).agg({'Рейтинг' : 'mean'})
print(df1)
# второй вариант   df.groupby(['Жанр', 'Год выпуска'])['Рейтинг'].mean()
```

```python
df1 = df.groupby('Жанр')['Рейтинг'].mean().reset_index().sort_values('Рейтинг', ascending = False)
print(df1)
df1 = df['Год выпуска'].value_counts().sort_index()
print(df1)
df1 = df.groupby(['Жанр' , 'Режиссер'])['Сборы (млн $)'].sum().reset_index().sort_values('Сборы (млн $)' , ascending = False)
print(df1)
```
