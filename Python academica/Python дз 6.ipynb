```python
def sum_of_two(num1, num2):
    return num1 + num2
sum_of_two(10, 5)

# 2 variant
# sum_of_two = lambda x,y: x+y
# print(sum_of_two(10,5))
```

```python
def is_even(num):
    if num % 2 == 0:
        return True
    else:
        return False
is_even(20)
# 2 variant 
# is_even = lambda x: x%2 == 0
```

```python
def find_max(*nums):
    return max(nums)
find_max(1,2,3,4,5)

# 2 variant
# find_max = lambda x: max(x)
# print(find_max([1,2,3,4,5])
```

```python
def c_f(x):
    return (x*1.8)+32
c_f(0)
# 2 variant
# c_f = lambda x: (x*1.8)+32
# print(c_f(0))
```

```python
def c_vowels(text):
    vowels = "аеёиоуыэюя"
    count = 0
    for i in text.lower():
        if i in vowels:
            count += 1
    return count
c_vowels("Привет")
```
