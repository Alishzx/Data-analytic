```python
class Users:
    def __init__(self, f_name, l_name, email, b_year,):
        self.f_name = f_name
        self.l_name = l_name
        self.email = email
        self.b_year = b_year
    def get_age(self):
        return 2023 - int(self.b_year)
users = []
try:
    file = open("users.txt", "r")
    for line in file:
        data = line.strip().split(",")
        user = Users(
            data[1],
            data[2],
            data[3],  
            data[4]
        )
        users.append(user)
    file.close()
except FileNotFoundError:
    print("the file that you would like to read is not found")
except Exception:
    print("some error occured")
def filter(users, low=30, high=50):
    for user in users:
        age = user.get_age()
        if low < age < high:
            print(user.email, age, sep=",")
filter(users)
```
