```python
height = int(input("Input height: "))
length = int(input("Input length: "))
for i in range(height):
    for j in range(length):
        print("*", end=" ")
    print()
```

```python
M = int(input("Input your number: "))
N = int(input("Input number to power: "))
for i in range (1, M+1):
    for j in range (1,N+1):
        print(f'{i}^{j} = {i**j}', end = " ")
    print()
```

```python
time = int(input())
while time != -1:
    print(f'{time}')
    time -= 1
print('Time is up!')
```

```python
import random
pc_num = random.randint(1, 100)
guest = 0

while guest != pc_num:
    guest = int(input('guess the number: '))
    if guest > pc_num:
        print('Your number is bigger: ')
    elif guest < pc_num:
        print('Your number is smaller: ')
    else:
        print('You guess the number!')
```
