```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def distance_to_origin(self):
        return (self.x ** 2 + self.y ** 2) ** 0.5
point1 = Point(10, 10)
print(point1.distance_to_origin())
```

```python
class Student:
    def __init__(self, name):
        self.name = name
        self.grades = []
    def add_grade(self, grade):
        self.grades.append(grade)
    def avg_grade(self):
        if len(self.grades) == 0:
            return 0
        return sum(self.grades) / len(self.grades)

student1 = Student('Alisher')
student1.add_grade(5)
student1.add_grade(4)
student1.add_grade(3)
print(student1.name)
print(student1.grades)
print(student1.avg_grade())
```

```python
class Book:
    def __init__(self, title, author, page):
        self.title = title
        self.author = author
        self.page = page
    def is_long(self):
        return self.page > 500

book1 = Book("Harry Potter", "Rowling", 700)
print(book1.is_long())
```
